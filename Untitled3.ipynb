```python
import pandas as pd #импорт библиотеки для работы с данными
import matplotlib.pyplot as plt  #библиотека, с помощью который будем строить график для оценки получившейся модели
from sklearn.model_selection import train_test_split #потребуется для разбиения данных на две категории: те, на которых
#наша модель будет обучаться и те, на которых мы будем ее проверять
from sklearn.linear_model import LogisticRegression #для нашей модели будем использовать метод логистической регрессии,
#так как он хорошо подходит для задач классификации и может быстро обрабатывать большие объёмы данных.
from sklearn.metrics import confusion_matrix # метод, который понадобится нам для оценки эффективности модели
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve # методы для построения графика
import re

df = pd.read_csv('data.csv') # считываем данные
x = df[[' Cash flow rate', ' Debt ratio %', ' Current Liabilities/Equity', ' Total income/Total expense', ' Interest Coverage Ratio (Interest expense to EBIT)', ' Net Income Flag']]
# Так как в таблице очень много различных признаков, я выбрал ключевые, по которым оценивают вероятность банкротства, чтобы не перенасыщать модель данными.
# Эти метрики я поместил в переменную x для дальнейшего использования в модели
y = df['Bankrupt?'][:] # в y хранятся метки, указывающие банкрот компания или нет

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42) # делим данные на обучающие и проверочные

logreg = LogisticRegression() # выбираем метод логистической регрессии
logreg.fit(x_train, y_train) # обучаем модель
y_pred = logreg.predict(x_test) # проверяем эффективность на тестовых данных
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(x_test, y_test)))

# Другой способ проверки через матрицу путаницы
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)

# Построение графика
# Мы будем строить ROC кривую для оценки модели, так как она хорошо подходит для задачи классификации
logit_roc_auc = roc_auc_score(y_test, logreg.predict(x_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(x_test)[:,1]) # задаем ROC кривую
#Оформляем график
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```
